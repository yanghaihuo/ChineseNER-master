```python
import pandas as pd
import os
import json
import sys
import time
import numpy as np
import os 


def formatrow(row, s_column, r_columns):
        s = row[s_column]
        if isinstance(s, dict):
            return [s.get(r_column) for r_column in r_columns]
        else :
            return [np.nan]*len(r_columns)

def fun(x,a):
    m=[]
    if isinstance(x, list):
        for i in x:
            if i:
                if isinstance(i, dict):
                    m.append(i.get(a))
        return m
    else:
        return np.nan

def fun1(x,a):
    m=[]
    
    if isinstance(x, float):
        return x
    elif isinstance(x, list):
        if x:
            return "&&&&&&".join('%s' %i for i in x)
        return x
    
    
os.chdir(r'F:\360MoveData\Users\Administrator\Desktop')
df = pd.read_excel(r'data.xlsx',sheet_name=1)     

result = []
for i, row in df.iterrows():
    colums = row.index.values
    d = {}
    for colum in colums:
        try:
            json_o = json.loads(row[colum])
            d[colum] = json_o
            label=json_o['type']
            colum_dest = colum
        except:
            d[colum] = row[colum]
    result.append(d)

df = pd.DataFrame(result)
df['data']=df[colum_dest].apply(lambda s: s.get('data') if isinstance(s, dict) else np.nan)
df = df.explode('data')
df['info1'] = df[colum_dest].apply(lambda x: x.get('info') if isinstance(x, dict) else np.nan)

if label=='TXT':
    lables = ['start', 'columnIndex', 'end','annotateType','info']
    df[lables] = df.apply(formatrow,axis=1,result_type="expand",args=('data', lables))   

    cols=[i for i in df.columns if i not in ['start', 'columnIndex', 'end','annotateType','info','data','info1']]
    print('转换列对应关系： ')
    for i, j in enumerate(cols):
        print(f'{i}\t{j}')
    dict_map = dict(enumerate(cols))

    while 1:
        m=input('请输入需要转换的序号： ').strip()
        if m.isdigit():
            split_field = int(m)
            if split_field>max(dict_map.keys()) or  split_field<min(dict_map.keys()):
                print('输入序号有误，请重新输入')
            else:
                break
        else:
            print('请输入数字格式')

    split_field = dict_map.get(split_field)


    df['ner']=df.apply(lambda x: str(x[split_field])[int(x['start']):int(x['end'])] if x['start']==x['start']   else np.nan ,axis=1)


elif label=='PHOTO':
    lables = ['shapeType', 'points','info']
    df[lables] = df.apply(formatrow,axis=1,result_type="expand",args=('data', lables))


df['data-info-val'] = df['info'].apply(lambda x: fun(x,'val'))
df['data-info-val-']=df['data-info-val'].apply(lambda x :fun1(x,'realLabelValue'))
df['info-val'] = df['info1'].apply(lambda x: fun(x,'val'))
df['info-val-']=df['info-val'].apply(lambda x :fun1(x,'realLabelValue'))


for temp in ['data-info-val-','info-val-']:
    m=df[temp].astype(str).str.split('&&&&&&',expand=True)
    m=m.applymap(lambda x: str(x).replace(" ",''))
    m=m.applymap(lambda x: x.replace("[{'realLabelValue':'",'').replace("'},{'realLabelValue':'",',').replace("'}]",'').replace("['",'').replace("']",'').replace("','",','))
    m.columns = [ temp + str(col+1)  if  str(col).isdigit() else col  for col in m.columns]
    df=pd.concat([df,m],axis=1)


df.drop(columns=[colum_dest,'data','info','info1','data-info-val','data-info-val-','info-val','info-val-'], inplace=True)
df = df.mask(df.applymap(str).eq('[]')|df.applymap(str).eq('[[]]')|df.applymap(str).eq('[[[]]]')|df.applymap(str).eq('nan'))
df.dropna(axis=1,how='all',inplace=True) 
df.reset_index(drop=True, inplace=True)
df
# df.to_excel('b.xlsx',index=False)
```
